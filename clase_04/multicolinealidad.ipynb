```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
from scipy import stats
```

### El objetivo de este notebook es facilitar la creación de regresiones con set de datos para evaluar hipótesis

```python
# Definimos los datos
X1 = [1, 2, 3, 4, 5]
#X2 = [10, 20, 30, 40, 50]
#X3 = [5, 4, 3, 2, 1]
Y  = [100, 200, 300, 400, 500]

# Creamos el DataFrame
df = pd.DataFrame({
    'X1': X1,
    #'X2': X2,
    #'X3': X3,
    'Y': Y
})

print(df)
```

```python
y=df['Y']
X=sm.add_constant(df.drop(columns='Y'))

reg_y_x=sm.OLS(y,X).fit()
display(reg_y_x.summary())
```

```python
y=df['X1']
X=sm.add_constant(df.drop(columns='X1'))

reg_x_y=sm.OLS(y,X).fit()
display(reg_x_y.summary())
```

### verificamos que multiplicar las pendientes de  las regresiones lineaes es igual al coeficiente de determinación de la regresión cruzada


```python
print(reg_x_y.params[1])
print(reg_y_x.params[1])
print(reg_x_y.rsquared)
print(reg_y_x.rsquared)

## verificamos que multiplicar las pendientes de  las regresiones lineaes es igual al coeficiente de determinación de la regresión cruzada
print(np.round(reg_x_y.params[1]*reg_y_x.params[1]))
```
