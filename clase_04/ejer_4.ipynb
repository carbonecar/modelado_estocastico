```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df_casa = pd.read_excel('./Ejemplo_Casa.xls')
display(df_casa.head())
```

```python
y=df_casa['PRECIO']
X=sm.add_constant(df_casa.drop(columns='PRECIO'))

reg_casa=sm.OLS(y,X).fit()
display(reg_casa.summary())
```

### Hacemos las modificaciones al df para quitar las observaciones, variables dummy, etc

```python
df_casa_filtrado=df_casa[df_casa['BANOS']!=4].copy()
y1=df_casa_filtrado['PRECIO']
X1=sm.add_constant(df_casa_filtrado.drop(columns='PRECIO'))
display(X1)
reg_casa_filtrado=sm.OLS(y1,X1).fit()
display(reg_casa_filtrado.summary())
```

```python
print(reg_casa_filtrado.params[0])
```

### b) Utilidad del estadístico F

El estadístico F sirve para ver si en su conjunto el modelo es significativo. La hipótesis nula es que todas las regresoras son =0  y la alternativa es que todas son distintas de cero. En este ejemplo el estadístico F es singnificativo lo que indica que el modelo globalmente esta bien, pero como hay coeficimentes no significativos, es posible que exista multicolinealidad.

Calculo: 

F=(RRSS-URSS)/q /(URSS/(1-k))

```python
print(f"F-statistic: {reg_casa_filtrado.fvalue}")
print(f"p-value del test F: {reg_casa_filtrado.f_pvalue}")
```

### El p-value es bajo, lo que implica que es muy significativo, por o tanto el modelo esta bien especificado en su conjunto y como hay esitmadores no signficativos, es posible que exista multicolinealidad.
