```python
import os
import pandas as pd
import yfinance as yf
import statsmodels.api as sm
import numpy as np
```

```python
btc_usd=yf.download('BTC-USD', start='2020-06-01', interval='1mo',auto_adjust=True)


btc_usd=btc_usd[['Close']].reset_index()

#Renombrar columnas
btc_usd.columns=['Date','btc-price']


#Calcular los retornos
btc_usd['btc_usd_ret'] = btc_usd['btc-price'].pct_change()

### graficamos la serie de los retornos
import matplotlib.pyplot as plt
btc_usd=btc_usd.dropna()
btc_usd.reset_index(drop=True)
plt.plot(btc_usd['Date'], btc_usd['btc_usd_ret'])
plt.title('Bitcoin Monthly Returns')
plt.xlabel('Date')
plt.ylabel('Returns')
plt.show() 


```
